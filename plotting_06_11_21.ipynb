We have created a variable called type1_value_counts. This variable countains the first 30 countires with the most players by origin. So it goes through our dataset called 'data' and into our column 'Nationality' and gets the value_counts() for those or the sum for each country and returns only the top 30 from most to least or in descending order. 

```python
type1_value_counts = data["Nationality"].value_counts()[0:30]
```

This is similar to our bar plot above except this visualisation gives us the top30 countries with the most players. 

```python
## use plt from libraries matplotlib.pyplot and get the figure with a figsize of (20,10)
plt.figure(figsize=(20,10))
## a variable called ax calls on the library squarify which plts the type1_value_counts 
ax = squarify.plot(type1_value_counts, 
## a variabled called label which gets the index of our type1_value-counts
              label=type1_value_counts.index,
## a variable called color, it calls on our library seaborn as sns and get the color_palette of 'tab20' and the length of type1_value_counts
              color=sns.color_palette("tab20", len(type1_value_counts)),
              pad=1,
              text_kwargs={'fontsize': 15})
## your variable ax and set the title for this visualization with the size =20
ax.set_title("Which countries have the most players?", size=20)
## this uses pyplot to exclude the axis 'x-axis and y-axis' in our visualisation as it is not required.
plt.axis('off');
```

This shows the histogram for the speciality scores of the players. 

And we can see that the highest score is around 1700 with about 900 players having that score. 

```python
# x contains our column Special
x = data.Special
# plt will use figure() to get the figsize 
plt.figure(figsize = (12, 8))
# plt will use the style tableau-colorblind10
plt.style.use('tableau-colorblind10')
# a variable ax is created which uses seaborn 'sns' to get a distpot with our variable x, with bins 'columns'=58, kde = false
ax = sns.distplot(x, bins = 58, kde = False, color = 'm')
# ax will set our label for the x-axis
ax.set_xlabel(xlabel = 'Special score range', fontsize = 16)
# ax will set the labeling for our y-axis
ax.set_ylabel(ylabel = 'Count of the Players',fontsize = 16)
# ax will set the label for the title 
ax.set_title(label = 'Histogram for the Speciality Scores of the Players', fontsize = 20)
plt.show()
```

Below is the histogram for the players potential scores.

This is a dataframe that goes through our dataset called data  iloc indexing, it will group the data by 'Position' and 'Overall' columns taking an idxmax() of that groupby. Then it conatain a list of the column to be included in our dataframe and chooses a style background as red for our column 'Age'. The darker the color, the older the player. 

This filters by position and shows which player is the best at that position looking at their overall score from our 'Overall' column. 

```python
# best players per each position with their age, club, and nationality based on their overall scores

data.iloc[data.groupby(data['Position'])['Overall'].idxmax()][['Position', 'Name','Age', 'Club', 'Nationality']].style.background_gradient('Reds')
```
